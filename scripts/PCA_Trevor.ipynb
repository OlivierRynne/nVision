```python
import numpy as np
import pandas as pd
import sklearn as skl
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.decomposition import PCA
```

```python
x1 = np.random.random(size=500)
x2 = 3 + x1*4 + 0.5*np.random.normal(size=500)
x3 = 5 - x1*3 + 4*x1**2 + 0.1*np.random.normal(size=500)
x4 = 2*np.random.random(size=500) -100
```

```python
plt.scatter(x1,x2, alpha=0.25)
```

```python
plt.scatter(x1,x3, alpha=0.25)
```

```python
plt.scatter(x1,x4, alpha=0.25)
```

```python
data = pd.DataFrame(data={0:x1, 1:x2, 2:x3, 3:x4})
```

```python
sns.pairplot(data)
```

```python
scaler = preprocessing.StandardScaler().fit(data)
```

```python
data_s = scaler.transform(data)
data_s = pd.DataFrame(data=data_s)
```

```python
sns.pairplot(data_s)
```

```python
pca = PCA(n_components=4)
data_pca = pca.fit(data_s).transform(data_s)
data_pca = pd.DataFrame(data=data_pca)
```

```python
sns.pairplot(data_pca)
```
